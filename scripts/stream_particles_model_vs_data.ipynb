### This notebook is where I'll be modelling the stream particles of Boo3, to compare with Kevin's data (plotted in V_r vs. RA, and proper motions).

```python
from galpy.orbit import Orbit
from galpy import potential
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
from matplotlib import animation
from IPython.display import HTML
from galpy.potential import MWPotential2014,ChandrasekharDynamicalFrictionForce,HernquistPotential,MovingObjectPotential
from galpy.util import conversion
from streamtools.df import streamspraydf

from stream_package.stream_package import graphics
```

```python
pot = MWPotential2014
tdisrupt=5
mass = 10e7

ro = 8.
vo = 220.
to = conversion.time_in_Gyr(ro=ro, vo=vo)
mo = conversion.mass_in_msol(ro=ro, vo=vo)

sixD_pace = [209.3, 26.8, 46.5, -1.176, -0.89, 197.5]
o = Orbit(vxvv = sixD_pace, ro=ro, vo=vo, solarmotion = [-11.1, 24.0, 7.25], radec=True)
o.turn_physical_off()
```

```python
%%time
o_lmc, oall_lmc, dtall_lmc = graphics.streamorbitslmc_acc(mass, o, tdisrupt, pot=pot, nstar=100)
```

```python
plt.scatter(oall_lmc.ra(),oall_lmc.vlos())
plt.xlim(207,213)
plt.ylim(165,230)
plt.xlabel('RA (deg)')
plt.ylabel('Radial velocity (km/s)')
```

```python
plt.scatter(oall_lmc.pmra(), oall_lmc.pmdec())
plt.xlim(2,0)
plt.ylim(-2,0)
plt.xlabel('PMRA (mas/yr)')
plt.ylabel('PMDEC (mas/yr)')
```

```python
o_lmc2, oall_lmc2, dtall_lmc2 = graphics.streamorbitslmc_acc(mass*5, o, tdisrupt, pot=pot, nstar=100)
```

```python
plt.scatter(oall_lmc2.ra(),oall_lmc2.vr())
plt.plot(o.ra(),o.vr(), 'ro')
#plt.xlim(207,213)
#plt.ylim(165,230)
plt.xlabel('RA (deg)')
plt.ylabel('Radial velocity (km/s)')
plt.show()

plt.scatter(oall_lmc2.pmra(), oall_lmc2.pmdec())
plt.xlim(2,0)
plt.ylim(-2,0)
plt.ylim(-2,0)
plt.xlabel('PMRA (mas/yr)')
plt.ylabel('PMDEC (mas/yr)')
plt.show()
```
