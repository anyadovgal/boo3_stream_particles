### Plot generation

```python
def gen_plot(oall):
    fig = plt.figure(figsize=(12,10))
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    ax3 = fig.add_subplot(223)
    ax4 = fig.add_subplot(224)
    
    ax1.scatter(oall.ra(), oall.dec())
    ax1.plot(o.ra(), o.dec(), "ro", label='Boo3')
    ax1.set_xlim(213, 206)
    ax1.set_ylim(24.5, 29)
    ax1.set_xlabel('RA (deg)')
    ax1.set_ylabel('DEC (deg)')
    ax1.legend()
    ax1.minorticks_on()
    
    ax2.scatter(oall.ra(), oall.vlos())
    ax2.plot(o.ra(), o.vlos(), "ro", label='Boo3')
    ax2.set_xlim(206, 213)
    ax2.set_ylim(165, 230)
    ax2.set_xlabel('RA (deg)')
    ax2.set_ylabel('Radial Velocity (km/s)')
    ax2.legend()
    ax2.minorticks_on()
    
    ax3.scatter(oall.ra(), oall.pmra())
    ax3.plot(o.ra(), o.pmra(), "ro", label='Boo3')
    ax3.set_xlim(206, 213)
    ax3.set_ylim(-2, 0)
    ax3.set_xlabel('RA (deg)')
    ax3.set_ylabel('PMRA (mas/yr)')
    ax3.legend()
    ax3.minorticks_on()
    
    ax4.scatter(oall.dec(), oall.pmdec())
    ax4.plot(o.dec(), o.pmdec(), "ro", label='Boo3')
    ax4.set_xlim(24.5, 29)
    ax4.set_ylim(-2, 0)
    ax4.set_xlabel('DEC (deg)')
    ax4.set_ylabel('PMDEC (mas/yr)')
    ax4.legend()
    ax4.minorticks_on()
    
    plt.tight_layout()
    plt.show();

    
    
    plt.scatter(oall.pmra(), oall.pmdec())
    plt.scatter(o.pmra(), o.pmdec(), c='red', label='Boo3')
    plt.legend()
    plt.xlim(-2,0)
    plt.ylim(-2,0)
    plt.xlabel('PMRA (mas/yr)')
    plt.ylabel('PMDEC (mas/yr)')
    ax = plt.gca()
    ax.minorticks_on()
    plt.show();
    return None
```

```python
for key, value in streams_dict.items():
    print(key)
    gen_plot(value)
```
