# Computing orbits of the MW satellite galaxies collection

## This notebook is going to use scipy to find the nearest pericentres/apocentres, rather than galpy's methods

When galpy calculates the pericentres and apocentres of orbits using o.rap() and o.rperi(), it uses a function similar to max() and min(), in which it just finds the smallest and largest values in the radial distance array. Usually this isn't a problem, but when orbits vary in period over time (due to perturbations, ex. MW acceleration caused by LMC inclusion) it can lead to too small pericentres & too large apocentres. Especially since these types of models get less accurate the further back in time you integrate, this can cause problems. If all of your bodies have similar periods then you can just reduce the integration time to remedy this, however if you're dealing with bodies where some of them end up having long periods & require long integration times (ex. dwarf galaxies), finding the nearest pericentre & apocentre can help remedy this problem.

```python
from galpy.orbit import Orbit
from galpy.util import conversion
from galpy.potential import MWPotential2014, ChandrasekharDynamicalFrictionForce,HernquistPotential,MovingObjectPotential
from galpy.potential.mwpotentials import McMillan17
import numpy as np
import matplotlib.pyplot as plt
```

## Set-up

```python
## Integration of galaxies:
tdisrupt=10 # Using a long integration time since some of the galaxies have long periods

# Setting McMillan17 as my potential
pot = McMillan17

ro = conversion.get_physical(pot)['ro']
vo = conversion.get_physical(pot)['vo']
to = conversion.time_in_Gyr(ro=ro, vo=vo)
mo = conversion.mass_in_msol(ro=ro, vo=vo)
ts = np.linspace(0, -tdisrupt/to, 1001)

# Initiating the orbit object for the collection
o_sats = Orbit.from_name('MW satellite galaxies', ro=ro, vo=vo)

# Integrating the orbit without LMC inclusion
o_sats.integrate(ts, pot)
```

### Set-up for LMC inclusion

***NOTE:*** If not using McMillan17 potential, in the ChandrasekharDynamicalFrictionForce line, use dens=pot rather than dens=pot[1]. pot[1] is only being used here because the McMillan17 potential model is too complicated and forces cannot be computed for the entire potential.

```python
mass_lmc=1.38e11 #solar masses
rscale_lmc=10.2 #kpc

pot = McMillan17

#Initialize and integrate the orbit of the LMC
#Note orbit has to be integrated back 10 Gyr
#Note we assume the LMC experienced dynamical friction due to MW
o_lmc = Orbit.from_name('LMC', ro=ro, vo=vo, solarmotion=[-11.1, 24.0, 7.25])
ts= np.linspace(0.,-tdisrupt/to,1001)
cdf= ChandrasekharDynamicalFrictionForce(GMs=mass_lmc/mo, rhm=rscale_lmc/ro, dens=pot[1], ro=ro,vo=vo)
o_lmc.integrate(ts,pot+cdf)

#Setup a moving Hernquist potential to represent the LMC
pot_lmc = HernquistPotential(mass_lmc/mo,rscale_lmc/ro,ro=ro,vo=vo)
moving_pot_lmc = MovingObjectPotential(o_lmc, pot_lmc,ro=ro,vo=vo)

#Add the moving Hernquest potential to the MW
total_pot = [pot]
total_pot += [moving_pot_lmc]
```

```python
from galpy.potential import (evaluateRforces, evaluatephitorques,
                                 evaluatezforces)
loc_origin= 1e-4 # Small offset in R to avoid numerical issues
ax= lambda t: evaluateRforces(moving_pot_lmc,loc_origin,0.,phi=0.,t=t,
                                use_physical=False)
ay= lambda t: evaluatephitorques(moving_pot_lmc,loc_origin,0.,phi=0.,t=t,
                                use_physical=False)/loc_origin
az= lambda t: evaluatezforces(moving_pot_lmc,loc_origin,0.,phi=0.,t=t,
                                use_physical=False)

t_intunits= o_lmc.time(use_physical=False)[::-1] # need to reverse the order for interp
ax4int= np.array([ax(t) for t in t_intunits])
ax_int= lambda t: np.interp(t,t_intunits,ax4int)
ay4int= np.array([ay(t) for t in t_intunits])
ay_int= lambda t: np.interp(t,t_intunits,ay4int)
az4int= np.array([az(t) for t in t_intunits])
az_int= lambda t: np.interp(t,t_intunits,az4int)

from galpy.potential import NonInertialFrameForce
nip= NonInertialFrameForce(a0=[ax_int,ay_int,az_int])
```

### Integrating for LMC inclusion

```python
# Keeping the LMC inclusion orbit separate from the exclusion orbit
o_sats_LMC = Orbit.from_name('MW satellite galaxies', ro=ro, vo=vo)
o_sats_LMC.integrate(ts, pot + moving_pot_lmc + nip)
```

Since the Pace et al. 2022 paper computed pericentre & apocentre data points, I'm going to compare the results from my galpy run with the Pace et al. results.

```python
pace_lmc_peri = data.pericenter.to_list()
pace_peri = data.pericenter_noLMC.to_list()
pace_lmc_ap = data.apocenter.to_list()
pace_ap = data.apocenter_noLMC.to_list()
names = data.name.to_list()

names = [s.replace(' ', '') for s in names] # since the galpy names dont have spaces
```

```python
# Find matches between both sets:
matches = list(set(names) & set(o_sats.name))
matches.sort()
```

```python
# Turn the matches into boolean arrays that can be applied on
#  the datasets, so that you are comparing apples to apples.

# Indices I can use on my generated galpy data
indm = np.array([], dtype=bool)
for i in range(len(o_sats.name)):
    if o_sats.name[i] in matches:
        indm = np.append(indm, True)
    else:
        indm = np.append(indm, False)

# Indices I can use on the Pace data
indd = np.array([], dtype=bool)
for i in range(len(names)):
    if names[i] in matches:
        indd = np.append(indd, True)
    else:
        indd = np.append(indd, False)
```

## Calculating the nearest pericentres & apocentres for the orbits I generated

The way that the code is written right now will attempt to set an orbit's pericentre and apocentre to be the nearest local minimum and maximum respectively. If it fails to do so (aka, if there is no local minimum or maximum), it will set this value to = 0, so that it is separated from the rest of the data set and can be examined individually. If you would like the pericentre to be the minimum value (likely will be today's radial distance), you can adjust the code as described in my comments. Similarily, if you would like the apocentre to be the to be the maxmimum value, you can use the same method, just with max() (or np.max()?) instead of min(). You can also do something like o[0] to get today's radial distance.

```python
from scipy.signal import argrelextrema

# Calculating the pericentres for orbits excluding LMC
sat_peri = []
for i in range(50):
    o = o_sats.r(ts)[i,:] # Get the radial distance array for orbit i, assign it to 'o'
    m_list = o[(argrelextrema(o, np.less)[0])] # Find all local minima
    if m_list.tolist() == []: # If there are no local minima
        sat_peri.append(0)    # Set the pericentre to 0, you can change this
                              #  to min(o) if you want the pericentre
                              #  to be the minimum.
    else: # If there is a local minimum
        sat_peri.append(m_list[0]) # Set orbit i's local minimum to be the nearest one
    
# Calculating the pericentres for orbits including LMC
sat_lmc_peri = []
for i in range(50):
    o = o_sats_LMC.r(ts)[i,:]
    m_list = o[(argrelextrema(o, np.less)[0])]
    if m_list.tolist() == []:
        sat_lmc_peri.append(0)   #min(o))
    else:
        sat_lmc_peri.append(m_list[0])
        
# Calculating the apocentres for orbits excluding LMC
sat_ap = []
for i in range(50):
    o = o_sats.r(ts)[i,:]
    m_list = o[(argrelextrema(o, np.greater)[0])]
    if m_list.tolist() == []:
        sat_ap.append(0)    #max(o))
    else:
        sat_ap.append(m_list[0])
        
# Calculating the apocentres for orbits including LMC
sat_lmc_ap = []
for i in range(50):
    o = o_sats_LMC.r(ts)[i,:]
    m_list = o[(argrelextrema(o, np.greater)[0])]
    if m_list.tolist() == []:
        sat_lmc_ap.append(0)    #max(o))
    else:
        sat_lmc_ap.append(m_list[0])
```
